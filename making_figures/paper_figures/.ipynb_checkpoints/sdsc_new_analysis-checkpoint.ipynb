Setting up analysis on SDSC

```python
import h5py as h5  #importing, make sure the kernel is correct or the module will be unknown
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
```

```python
# reading in the HDF5 file, try opening loop 
pathToweights_H5 = '/expanse/lustre/scratch/msantiago/temp_project/COMPAS_DATA/N_5e6_MassiveWDWD_NSNS_fid/MainRun/COMPAS_Output_wWeights.h5' #path of the hdf5 file

Data_weights = h5.File(pathToweights_H5, "r")

Data_weights.keys()
```

```python
DCOs = Data_weights['BSE_Double_Compact_Objects'] # gathering the DCO group
DCOs.keys()
```

```python
col_times = DCOs['Coalescence_Time'][()]
lifetimes = DCOs['Time'][()]
```

```python
col_times
```

```python
plt.hist(lifetimes)
plt.xlabel("Time (Myr)")
```

```python
plt.hist(np.log10(col_times))
plt.xlabel("Coalescence Time (Myr) in log")
```

```python
delay_times = col_times + lifetimes
```

```python
plt.hist(np.log(delay_times))
plt.xlabel("Delay Time (Myr) in log")
plt.axvline(np.log(13.8e3), color='r',linestyle='--')
```

```python
DCO_mask = Data_weights['Rates_mu00.025_muz-0.049_alpha-1.79_sigma01.129_sigmaz0.048']['DCOmask'][()]

# let's mask our data and making sure it only picked out those with delay times smaller than age of unvierse
col_times_mask = col_times[DCO_mask]
lifetimes_mask = lifetimes[DCO_mask]
delay_times_mask = col_times_mask + lifetimes_mask
```

```python
plt.hist(np.log(delay_times_mask))
plt.xlabel("Delay Time (Myr) in log")
plt.axvline(np.log(13.8e3), color='r',linestyle='--')
```
