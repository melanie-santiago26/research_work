```python
# let's first look at the NSNS_output

# the average_SF_mass_needed is Average_SF_mass_needed =  10892909.034667056

pathToH5_NSNS = '/expanse/lustre/scratch/msantiago/temp_project/COMPAS_DATA/N5e6_NSNS_CE_alpha2/MainRun/COMPAS_Output_wWeights.h5'

Data_NSNS  = h5.File(pathToH5_NSNS, "r")

DCOs_NSNS = Data_NSNS['BSE_Double_Compact_Objects'] # getting the DCO objects

# gathering the double compact objects that we have computed rates for
DCO_mask_NSNS = Data_NSNS['Rates_mu00.025_muz-0.049_alpha-1.79_sigma01.129_sigmaz0.048']['DCOmask'][()]

# making sure we are rpesenting the true universe
mixture_weights_all = DCOs_NSNS['mixture_weight'][()]
mixture_weights = mixture_weights_all[DCO_mask_NSNS]

# times (these should be in Myr)
lifetimes_all = DCOs_NSNS['Time'][()]
lifetimes = lifetimes_all[DCO_mask_NSNS]

col_times_all = DCOs_NSNS['Coalescence_Time'][()]
col_times = col_times_all[DCO_mask_NSNS]

# this should hypothetically be the DCO systems that merge within a hubble time (since DCO mask was applied) 
delay_times = lifetimes + col_times
```

```python
## all NSNS delay time dist of all systems
age_universe = 13.9e9
delay_time_NSNS_log = np.log10(delay_times*(1e6))
hist_NSNS, bin_edges_NSNS = np.histogram(delay_time_NSNS_log, weights=mixture_weights, bins=np.linspace(7.5,np.log10(age_universe),20))
center_bins_NSNS = (bin_edges_NSNS[:-1] + bin_edges_NSNS[1:])/2
bin_width_NSNS = np.diff(bin_edges_NSNS)

plt.plot(center_bins_NSNS,(hist_NSNS/bin_width_NSNS)*1e-3,color='darkorchid',lw=2, label=r'$\mathrm{t_{delay}}$')




plt.text(7, 45, r'$\times 10^3$', fontsize = 15)

# ax[0].set_xlabel(r"$\log_{10}$(t/$\mathrm{Yr}$)",fontsize=15)
# plt.yscale('log')
# plt.xlim(0,15e9)
# plt.ylim(1e-1,1e3)
plt.ylabel(r"$\mathrm{dN}$/$\mathrm{d}\log_{10}(t$)",fontsize=15) 
plt.xlabel(r"$\log_{10}(t$/$\mathrm{Yr}$)", fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.axvline(np.log10(age_universe), color='pink', linestyle='--', linewidth=2)#,label='Hubble Time')
# plt.legend(fontsize=15)
# plt.title("NSNS", fontsize = 20)
```

```python
# let's see if there are any system that merge beyond a hubble time

hubble_time = 13.9e3
condition_no_mergers = delay_times > hubble_time

no_mergers = np.sum(condition_no_mergers)

merged_systems = len(delay_times) - no_mergers

condition_mergers = delay_times < hubble_time
mergers = np.sum(condition_mergers)

print("The number of systems that merge within a hubble time is = {}." .format(merged_systems))
print("The number of systems that merge withina  hubble time is = {}" .format(mergers))

# yay both ways get the same value!
```

```python
# let's look at what types of systems these are

stellar_types_1_all = DCOs_NSNS['Stellar_Type(1)'][()]
stellar_types_1 = stellar_types_1_all[DCO_mask_NSNS]
stellar_types_1_merged = stellar_types_1[condition_mergers]

stellar_types_2_all = DCOs_NSNS['Stellar_Type(2)'][()]
stellar_types_2 = stellar_types_2_all[DCO_mask_NSNS]
stellar_types_2_merged = stellar_types_2[condition_mergers]

plt.plot(stellar_types_1_merged,stellar_types_2_merged,'.')
```

```python
# let's find the rate of NSNS mergers
rates_DCO = Data_NSNS['Rates_mu00.025_muz-0.049_alpha-1.79_sigma01.129_sigmaz0.048']['merger_rate'][()]
rates_DCO_masked = rates_DCO[condition_mergers]

NSNS_systems_bool = np.logical_and(stellar_types_1_merged==13, stellar_types_2_merged==13)
NSNS_rate = np.sum(rates_DCO_masked[NSNS_systems_bool], axis=0)
```

```python
## all NSNS delay time dist of all systems
age_universe = 13.9e9
delay_time_NSNS_log = np.log10(delay_times*(1e6))
hist_NSNS, bin_edges_NSNS = np.histogram(delay_time_NSNS_log, weights=mixture_weights, bins=np.linspace(7.5,np.log10(age_universe),20))
center_bins_NSNS = (bin_edges_NSNS[:-1] + bin_edges_NSNS[1:])/2
bin_width_NSNS = np.diff(bin_edges_NSNS)

plt.plot(center_bins_NSNS,(hist_NSNS/bin_width_NSNS)*1e-3,color='darkorchid',lw=2, label=r'$\mathrm{t_{delay}}$')




plt.text(7, 45, r'$\times 10^3$', fontsize = 15)

# ax[0].set_xlabel(r"$\log_{10}$(t/$\mathrm{Yr}$)",fontsize=15)
# plt.yscale('log')
# plt.xlim(0,15e9)
# plt.ylim(1e-1,1e3)
plt.ylabel(r"$\mathrm{dN}$/$\mathrm{d}\log_{10}(t$)",fontsize=15) 
plt.xlabel(r"$\log_{10}(t$/$\mathrm{Yr}$)", fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.axvline(np.log10(age_universe), color='pink', linestyle='--', linewidth=2)#,label='Hubble Time')
# plt.legend(fontsize=15)
# plt.title("NSNS", fontsize = 20)
```
